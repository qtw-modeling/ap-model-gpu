```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function_periodic(t, T):
    
    ''' We conisder only t >= 0! '''
    
    offset = (t // float(T)) * T
    
    def func_single_period(t, T):
        if t < T/2.:
            return 0.
        else:
            return 1.
        
    return func_single_period(t - offset, T)
```

```python
T = 2.

# for plotting
xData = np.linspace(0, 10, 500)
yData = [function_periodic(x, T) for x in xData]

plt.figure(figsize=(11, 6))
plt.plot(xData, yData, lw=2)
plt.show()
```
